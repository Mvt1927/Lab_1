```python
# Question 2 Create vectors and try all of the similarity distances
from math import *
from decimal import Decimal


class Similarity():
    """ Five similarity measures function """

    def euclidean_distance(self, x: list, y: list):
        """ Return euclidean distance between two lists.

        sử dụng công thức `sqrt(Sigma(pow(abs(xi-yi),2)))`.

        Args:
            `x (list)`: a list store x value
            `y (list)`: a list store y value

        Returns:
            `float` : euclidean distance between two lists 
            `-1`    : error return -1,
        """

        # Kiểm tra size của 2 list đưa vào nếu sai trả về -1
        sizeX = len(x)
        sizeY = len(y)
        if sizeX != sizeY:
            return -1

        # Khởi tạo Sigma
        Sigma = sum(pow(abs(xi-yi),2) for xi,yi in zip(x,y))

        # Trả về distance = sqrt(Sigma)
        return sqrt(Sigma)

    def manhattan_distance(self, x: list, y: list):
        """ Return manhattan distance between two lists.

        Sử dụng công thức `Sigma(abs(xi-yi))`

        Args:
            `x (list)`: a list store x value
            `y (list)`: a list store y value

        Returns:
            `float` : euclidean distance between two lists 
            `-1`    : error return -1,
        """

        # Kiểm tra size của 2 list đưa vào nếu sai trả về -1
        sizeX = len(x)
        sizeY = len(y)
        if sizeX != sizeY:
            return -1

        Sigma = sum(abs(xi-yi) for xi, yi in zip (x,y))

        # Trả về distance = Sigma
        return Sigma

    def minkowski_distance(self, x: list, y: list, p_value: float):
        """ return minkowski distance between two lists """
        return self.nth_root(sum(pow(abs(a-b), p_value) for a, b in zip(x, y)), p_value)

    def nth_root(self, value, n_root):
        """ returns the n_root of an value """
        root_value = 1/float(n_root)
        return round(Decimal(value) ** Decimal(root_value), 3)

    def cosine_similarity(self, x:list, y:list):
        """ Return cosine similarity between two lists
        
        sử dụng công thức `Sigma(xi*yi)/(sqrt(Sigma(pow(xi,2)))*sqrt(Sigma(pow(yi,2))))`
        
        Args:
            `x (list)`: a list store x value
            `y (list)`: a list store y value

        Returns:
            `float`: cosine similarity between two lists.
            `None`: Error
        """
        
        # Kiểm tra size của 2 list đưa vào nếu sai trả về None
        sizeX = len(x)
        sizeY = len(y)
        if sizeX != sizeY:
            return None
        
        SigmaXiYi = 0
        SigmaXi2 = 0
        SigmaYi2 = 0

        for xi,yi in zip(x,y):
            SigmaXiYi = SigmaXiYi + xi*yi
            SigmaXi2 = SigmaXi2 + pow(xi,2)
            SigmaYi2 = SigmaYi2 + pow(yi,2)
            
        return SigmaXiYi/(sqrt(SigmaXi2)*sqrt(SigmaYi2))
    
    def square_rooted(self, x):
        """ return 3 rounded square rooted value """
        return round(sqrt(sum([a*a for a in x])), 3)

    def jaccard_similarity(self, x, y):
        """ returns the jaccard similarity between two lists """
        intersection_cardinality = len(set.intersection(*[set(x), set(y)]))
        union_cardinality = len(set.union(*[set(x), set(y)]))
        return intersection_cardinality/float(union_cardinality)
```

```python
def main():
    """ the main function to create Similarity class instance and get used to it """
    measures = Similarity()
    a = [0,3,4,5]
    b = [7,6,3,-1]
    c = [0,1,2,5,6]
    d = [0,2,3,5,7,9]
    
    print("Euclidean : " ,measures.euclidean_distance(a,b))
    print("Manhattan : " ,measures.manhattan_distance(a,b))
    print("Minkowski : " ,measures.minkowski_distance(a,b, 0.5))
    print("cosine_similarity : " ,measures.cosine_similarity(a,b))
    print("Jaccard : " ,measures.jaccard_similarity(c,d))
 
if __name__ == "__main__":
    main()
```
